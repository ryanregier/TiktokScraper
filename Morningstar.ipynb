```python
import pandas as pd
import numpy as np 
import datetime
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.linear_model import LinearRegression
```

```python
investing_videos = pd.read_csv("#investing.csv")
younginvestor_videos = pd.read_csv("#younginvestor.csv")
finance_videos = pd.read_csv("#finance.csv")
```

```python
to_drop=['id', 'secretID', 'authorMeta.id', 'authorMeta.nickName', 'authorMeta.following', 'authorMeta.verified', 'authorMeta.signature',
       'authorMeta.avatar', 'musicMeta.musicId', 'musicMeta.musicOriginal', 'musicMeta.playUrl',
       'covers.default', 'covers.origin', 'covers.dynamic',
       'webVideoUrl', 'videoUrl', 'videoUrlNoWaterMark', 'videoMeta.width',
       'videoMeta.height',  'downloaded', 'authorMeta.secUid', 'musicMeta.musicName', 'musicMeta.duration', 
         'videoApiUrlNoWaterMark', 'musicMeta.musicAlbum', 'musicMeta.coverMedium', 'musicMeta.coverLarge','effectStickers']
```

```python
investing_videos = investing_videos.drop(to_drop, axis=1)
younginvestor_videos = younginvestor_videos.drop(to_drop, axis=1)
finance_videos = finance_videos.drop(to_drop, axis=1)

hashtag_dfs = [investing_videos, younginvestor_videos, finance_videos]
```

```python
for df in hashtag_dfs:
    df["Engagement Rates"] = ((df["diggCount"] + df["commentCount"] + df["shareCount"]) / df["playCount"]) * 100
```
